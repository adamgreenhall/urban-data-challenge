```python
import pandas as pd
from utils import df_to_json, json
```

```python
store = pd.HDFStore('data/san-francisco/data.hd5')
store
```

```python
stops = store['stops']
schedule = store['schedule']
pcounts = store['passenger_counts']
```

# Stops map

```python
stops = store['stops']
# get the city's center for a map 
print(stops[['latitude', 'longitude']].mean().tolist())

# get the bounds
print(stops[['latitude', 'longitude']].min().tolist())
print(stops[['latitude', 'longitude']].max().tolist())
```

```python
df_to_json(stops, 'web/stops_sf.json');
```

# Routes

```python
routes = []
for route, route_stops in stops.reset_index().groupby('id_route').groups.iteritems():
    routes.append({'route': str(route),
         'stops': df_to_json(
            stops.ix[route_stops].sort(['DIR', 'stop_seq']).reset_index().set_index(['DIR', 'id_stop'])
          )
    })
with open('web/routes_sf.json', 'w+') as f: f.write(json.dumps(routes))
```
