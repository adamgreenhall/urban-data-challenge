```python
def get_muni_route_number(rnm):
    # translate route_name into id_route
    # 5xx = xx limited route, 6xx = owl routes, 7xx = xx express route, 8xx = xx BX express route, 9xx = xx AX express route
    rid = rnm = str(rnm)
    if 'AX' in rnm:
        rid = '9' + rnm.strip('AX')
    elif 'BX' in rnm:
        rid = '8' + rnm.strip('BX')
    elif 'X' in rnm:
        rid = '7' + rnm.strip('X')
    elif 'OWL' in rnm:
        rid = '6' + rnm.strip(' OWL')
    elif rnm.endswith('L'):
        rid = '5' + rnm.strip('L')
    
    try:
        return int(rid.lstrip('0'))
    except ValueError:
        return None
```

## Parse just the malformed date lines

this is going to be a pain

```python
good_lines = pd.Index(range(513786)).diff(bad_lines).drop(0)
df = pd.read_csv('data/san-francisco/passenger-count.csv',
    skiprows=good_lines,
    usecols=usecols,
    parse_dates=[
        ['MO', 'DAY', 'YR'],
        ]
    ).rename(columns=rename_cols)
df
```

```python
timecols = ['TIMESTOP', 'TIMEDOORCLOSE', 'TIMEPULLOUT']

dfbt = df.copy()

# yuck!
for col in timecols:
    hrcol = 'hour_' + col
    datecol = 'date_' + col
    dfbt[datecol] = dfbt.date.copy().astype(str)
    dfbt[hrcol] = dfbt[col].apply(lambda ts: ts.split(':')[0]).astype(int)
    
    # if the hour is more than 23 (invalid), subtract 24hrs from the hour
    dfbt[col][ dfbt[hrcol] > 23 ] = dfbt[col][ dfbt[hrcol] > 23 ].apply(
        lambda ts: '{h:02d}:{mmss}'.format(h=int(ts.split(':')[0]) - 24, mmss=ts.split(':', 1)[1])
    )
    # and add one day to the date
    dfbt[datecol][ dfbt[hrcol] > 23 ] = dfbt[datecol][ dfbt[hrcol] > 23 ].apply(lambda date: str(pd.Timestamp(date) + pd.DateOffset(days=1)))
    
    # put the date and time back together
    dfbt[col] = (dfbt[datecol].apply(lambda ts: ts.split(' ')[0].split('T')[0]) + ' ' +  dfbt[col]).apply(lambda t: pd.Timestamp(t))

dfbt
```

```python
dfbt_clean = dfbt.drop(['date_' + col for col in timecols] + ['hour_' + col for col in timecols], axis=1).rename(columns=dict(
    TIMESTOP='time_stop',
    TIMEDOORCLOSE='time_door_close',
    TIMEPULLOUT='time_pullout',
))

dfbt_clean
```

# topojson

### stops

```python
with open('data/san-francisco/topojson/stops.json', 'r') as f: 
    stop_data = json.loads(f.read())
    
replacements = {
    'STOPID': 'id_stop',
    'STOPNAME': 'name_stop',
    }

def nice_stop_name(name):
    return name.replace('&', ' & ').rsplit(' ', 1)[0]

stop_df = pd.DataFrame(map(lambda obj: obj['properties'], stop_data['objects']['stops']['geometries']))
stop_df = stop_df.rename(columns=replacements)[replacements.values()]
stop_df['name_stop'] = stop_df.name_stop.apply(nice_stop_name)

stop_df.head()
```

```python
for i, obj in enumerate(stop_data['objects']['stops']['geometries']):
    obj['properties'] = stop_df.ix[i].apply(utils.native_types).to_dict()

with open('web/data/san-francisco/stops.json', 'w+') as f:
    f.write(json.dumps(stop_data))
```

## routes

```python
with open('data/san-francisco/topojson/routes.json', 'r') as f: 
    route_data = json.loads(f.read())

# routes paths can have multiple IDs
# -- but only one route can be selected per click
# -- so make the id_route the most likely one (simplest, and in valid_routes_pcounts)

route_df = pd.DataFrame(map(
    lambda obj: obj['properties'], route_data['objects']['routes']['geometries'])
    ).rename(columns=dict(LINENAME='name_route'))

route_df['day_of_week'] = ''
route_df.ix[route_df.LINEABBR.str.endswith('WKY'), 'day_of_week'] = 'weekday'
route_df.ix[route_df.LINEABBR.str.endswith('SAT'), 'day_of_week'] = 'saturday'
route_df = route_df[['name_route', 'day_of_week']]

def parse_route_ids(full_name):
    route_names = full_name.replace(' OWL', 'OWL').replace('/', '-').split(' ')[0].split('-')
    route_ids = map(get_muni_route_number, route_names)
    route_ids = [int(n) for n in pd.Index(route_ids).intersection(valid_routes_pcounts).tolist()]
    if pd.Series(route_ids).sum() == 0:
        route_ids = []
    return route_ids
def most_likely_id(route_ids):
    if len(route_ids) == 0: return None
    else: return route_ids[0]

route_df['all_route_ids'] = route_df.name_route.apply(parse_route_ids)
    
route_df['id_route'] = route_df.all_route_ids.apply(most_likely_id)

# filter to only routes which have defined ids
route_df = route_df[pd.notnull(route_df.id_route)]\
    [['id_route', 'name_route', 'all_route_ids', 'day_of_week']]

route_df['id_route'] = route_df.id_route.astype(int)
    
route_df
```

```python
# filter the json routes to only ones with valid id_route
route_data['objects']['routes']['geometries'] = \
    [route_data['objects']['routes']['geometries'][i] for i in route_df.index]

# redifine the topojson properties
for i, obj in enumerate(route_data['objects']['routes']['geometries']):
    obj['properties'] = route_df.reset_index().ix[i].apply(utils.native_types).to_dict()
    
with open('web/data/san-francisco/routes.json', 'w+') as f:
    f.write(json.dumps(route_data))
```
