# Get id_stop

It turns out that the real-time arrivals data has id_stop

```python
arrivals = pd.read_csv('data/san-francisco/realtime-arrivals.csv',
    usecols=[
        'PUBLIC_ROUTE_NAME',
        'TRIP_ID',
        'NEXTBUS_ARRIVAL_TIME',
        'LATITUDE',
        'LONGITUDE',
        'STOP_ID',
        ],
    parse_dates=['NEXTBUS_ARRIVAL_TIME']
    ).rename(columns=dict(
        PUBLIC_ROUTE_NAME='route_name',
        TRIP_ID='id_trip',
        NEXTBUS_ARRIVAL_TIME='time_arrival',
        LATITUDE='latitude',
        LONGITUDE='longitude',
        STOP_ID='id_stop',
    ))

store['arrivals'] = arrivals
arrivals
```

```python
schedule['date'] = schedule.time_scheduled_arrival.apply(lambda t: t.date())
arrivals['date'] = arrivals.time_arrival.apply(lambda t: t.date())
```

```python
sched_arrivals = pd.merge(arrivals, schedule, 
    on=['route_name', 'id_trip', 'latitude', 'longitude', 'date'],
    how='inner')
sched_arrivals
```

```python
# train lines still don't have route ids
sched_arrivals.route_name[pd.isnull(sched_arrivals.id_route)].unique()
```

```python
# assume anything more than 4hrs late is bad data
sched_arrivals = sched_arrivals[
    (sched_arrivals.time_arrival - sched_arrivals.time_scheduled_arrival).apply(
    lambda t: np.abs(t.total_seconds())
    ) < 4*3600]

# drop the rows without a route id and convert to integer
sched_arrivals = sched_arrivals[sched_arrivals.id_route.notnull()]
sched_arrivals['id_route'] = sched_arrivals.id_route.astype(int)
```

```python
sched_arrivals
```

```python
store['scheduled_arrivals'] = sched_arrivals
```

# Get id_trip

```python
# label the trip by the first scheduled time of the trip
sched_arrivals['id_trip_start_time_stop'] = ''

for idx, grp in sched_arrivals.sort(['date', 'id_route', 'id_trip', 'time_scheduled_arrival']).groupby(
    ['date', 'id_route', 'id_trip']):
    t = grp.ix[grp.index[0], 'time_scheduled_arrival']
    sched_arrivals.ix[grp.index, 'id_trip_start_time_stop'] = "{h}{m}_{sid}".format(
        h=t.hour, m=str(t.minute).zfill(2), 
        sid=grp.ix[grp.index[0], 'id_stop'])
```

```python
pcounts['date'] = pcounts.date.apply(lambda t: t.date())
```

```python
# pcounts['id_trip_start_time_stop'] = pcounts.id_trip.astype('str') + '_' + pcounts.id_stop.astype('str')

# label the passenger counts the same way (except the id_trip is already setup in HMM form)
pcounts['id_trip_start_time_stop'] = ''

for idx, grp in pcounts.sort(['date', 'id_route', 'id_trip', 'time_stop']).groupby(
    ['date', 'id_route', 'id_trip']):
    pcounts.ix[grp.index, 'id_trip_start_time_stop'] = "{tid}_{sid}".format(
        tid=grp.ix[grp.index[0], 'id_trip'], 
        sid=grp.ix[grp.index[0], 'id_stop'])
    
# sched_arrivals['id_trip_pcount'] = sched_arrivals.id_trip_pcount.astype(int)
```

```python
sched_pcounts = pd.merge(
    sched_arrivals.drop(['id_trip', 'latitude', 'longitude', ], axis=1), #.rename(columns={'id_trip_pcount': 'id_trip'}),
    pcounts, 
    left_on=['date', 'id_route', 'id_trip_start_time_stop', 'id_stop'],
    right_on=['date', 'id_route',  'id_trip_start_time_stop', 'id_stop'],
    how='inner'
)
sched_pcounts
```

```python
store['counts_w_schedule'] = sched_pcounts.drop(
        ['id_trip_start_time_stop', 'id_trip_pcount', 'route_name'], axis=1
    ).sort(
        ['date', 'id_route', 'id_trip', 'time_scheduled_arrival']
    ).set_index(
        ['date', 'id_route', 'id_trip', 'id_stop'])
```
